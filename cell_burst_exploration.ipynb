```python
import aligater as ag
```

```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
fcslist=ag.collectFiles("/media/ludvig/Project_Storage/BloodVariome/data/Phase_I/FCS/01-03-2016/plate 1/",lFilter=['General_'])
```

```python
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]
```

```python
for fcs_file in fcslist:
    sample=ag.loadFCS(fcs_file,return_type="agsample",compensate=False)
    #Collect time data
    time=sample.fcsDF['Time']
    #How many events are there in total?
    nOfEvents=len(time)
    #Check difference in time between adjacent events, in order time 0->last
    event_diff=[]
    vI=[]
    for n in np.arange(0,nOfEvents-1,1):
        event_diff.append(time[n+1]-time[n])
        vI.append(n)
    #split time into 10 000 event chunks and calculate the mean time for these events
    chunk_mean=[]
    for chunk in list(chunks(time,1000)):
        chunk_mean.append(np.mean(chunk))
    #Calculate the difference between these chunks
    diff_chunk_means=np.diff(chunk_mean)
    
    fig, axs = plt.subplots(3, 1, constrained_layout=True)
    #plot histogram number of events/time
    axs[0].set_title('Histogram event time (50 bins)')
    n, edges, patches=axs[0].hist(time,bins=50)
    axs[0].set_xlabel('Time')
    axs[0].set_ylabel('Frequency')
    #plot neighboor event diff
    axs[1].set_title('Neighbour event difference')
    axs[1].plot(time[:-1],event_diff)
    axs[1].set_xlabel('Time')
    axs[1].set_ylabel('Time-diff')
    #plot mean chunk difference over mean chunks
    axs[2].set_title('Neighbouring chunk time difference (1000 events/chunk)')
    axs[2].plot(chunk_mean[:-1],diff_chunk_means)
    axs[2].set_xlabel('Time')
    axs[2].set_ylabel('Time-diff')
    ymin, ymax = axs[2].get_ylim()
    ymin=0
    if ymax < 50:
        ymax = 50
    axs[2].set_ylim([ymin,ymax])
    plt.show(fig)
    plt.close(fig)
```
